```python
import collections 
d = collections.deque('abcdefg')
d.append('h')
print(d)
d.appendleft('1')
print(d)
d.pop()
print(d)
d.popleft()
print(d)
```

```python
import collections 
import threading

d = collections.deque()
def producer():
    for i in range(1000):
        d.append(i)
        
        
def comsumer(i):
    while True:
        try:
            num = d.pop()
            print("线程{} 正在执行任务{}".format(i,num))
        except:
            break

producer()        
ts = [threading.Thread(target=comsumer,args=(i,))for i in range(10)]
for t in ts:
    t.start()

for t in ts:
    t.join()
    
        
```
